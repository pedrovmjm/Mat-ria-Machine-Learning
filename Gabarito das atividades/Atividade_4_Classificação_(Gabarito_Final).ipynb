### Exercício 1.2 [3 pontos]

Existem diversos tipos de métricas de desempenho e é essencial que possamos entender a diferença entre elas e suas relações. Explique, com suas palavras as seguintes métricas e sua importância:

(a) Acurácia, Precisão, Cobertura, Sensitividade e Especificidade. [1 ponto]

***Resposta:***

**Acurácia:** indica quantos exemplos o modelo acertou em relação ao total de exemplos, sendo uma métrica importante para a avaliação de desempenho dos classificadores. 

$Acuracia = \frac{Verdadeiros Positivos + Verdadeiros Negativos}{Todas as amostras}$

**Precisão:** indica a porcentagem dos valores que realmente são positivos dentre os valores indicados como positivo pelo classificador (taxa de verdadeiros positivos), sendo uma métrica importante para a avaliação de desempenho dos classificadores, por representar a porcentagem de resultados positivos relevantes.

$Precisao = \frac{Verdadeiros Positivos}{Verdadeiros Positivos + Falsos Positivos}$

**Cobertura:** indica a porcentagem dos valores positivos que foram corretamente capturados pelo classificador, sendo uma métrica importante para a avaliação de desempenho dos classificadores, por representar a porcentagem de itens relevantes que foram selecionados. 

$Cobertura = \frac{Verdadeiros Positivos}{Verdadeiros Positivos + Falsos Negativos}$

**Sensitividade:** é o mesmo que cobertura. O termo cobertura é mais frequente na litetura de aprendizado de máquina, enquanto o termo sensitividade é mais frequente na comunidade estatística e na área da saúde.

$Sensitividade = \frac{Verdadeiros Positivos}{Verdadeiros Positivos + Falsos Negativos}$

**Especificidade:** indica a porcentagem dos valores que realmente são negativos dentre os valores indicados como negativo pelo classificador (taxa de verdadeiros negativos), sendo uma métrica importante para a avaliação de desempenho dos classificadores, por representar a porcentagem de resultados negativos relevantes que foram selecionados.

$Especificidade = \frac{Verdadeiros Negativos}{Verdadeiros Negativos + Falsos Positivos}$

(c) O que é a Matriz de Confusão? Cite 2 exemplos onde pode ser útil. [0.5 ponto]

**Resposta:**

A matriz de confusão é uma matriz tabela, na qual as colunas representam as classes atuais (resultados esperados/verdadeiros) e as linhas a saída do classificador (resultados do modelo). A soma dos elementos da diagonal principal dessa matriz dividida pela soma de todos os elementos da matriz dá a acurácia do modelo. A divisão do elemento da diagonal principal que está em uma linha, por todos os elementos desta linha dá a precisão do modelo e a divisão do elementos da diagonal principal que está em uma coluna, por todos os elementos desta coluna dá a cobertura.

A matriz de confusão é uma métrica útil em todos os exemplos citados acima, já que ela fornece um panorama geral sobre as métricas discutidas.

```python
# Realização da classificação com LR
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis

clf  = LogisticRegression(random_state=0, solver='newton-cg')
clda = LinearDiscriminantAnalysis()
cqda = QuadraticDiscriminantAnalysis()
clf  = clf.fit(df,y)
clda = clda.fit(df,y)
cqda = cqda.fit(df,y)
```

```python
# Imprimindo métricas de desempenho (no caso acurácia, precisão e cobertura)
from sklearn.metrics import *
# acc  = accuracy_score(y, pred)
pred1 = clf.predict(df)
prec = precision_score(y, pred1)
rec  = recall_score(y, pred1)
print('LR : prec=%.2f, rec=%.2f'%(prec,rec))

pred2 = clda.predict(df)
prec = precision_score(y, pred2)
rec  = recall_score(y, pred2)
print('LDA: prec=%.2f, rec=%.2f'%(prec,rec))

pred3 = cqda.predict(df)
prec = precision_score(y, pred3)
rec  = recall_score(y, pred3)
print('QDA: prec=%.2f, rec=%.2f'%(prec,rec))
```

```python
# Imprimindo a Matriz de Confusão
print("LR")
print(confusion_matrix(y,pred1))
print("LDA")
print(confusion_matrix(y,pred2))
print("QDA")
print(confusion_matrix(y,pred3))
```

2. Discuta os resultados, avaliando como as probabilidades mudaram as métricas medidas.

```python
#  Definição do limiar para observar mudanças nos resultados.
def evaluate_model(clf, df, y,limiar=-1):

  if limiar >= 0:
    pred_proba = clf.predict_proba(df)
    pred       = (pred_proba[:,1]>limiar)*1
  else:
    pred = clf.predict(df)
    limiar = 0.5

  acc  = accuracy_score(y, pred)
  prec = precision_score(y, pred)
  rec  = recall_score(y, pred)
  print('limiar=%.3f acc=%.3f, prec=%.3f, rec=%.3f'%(limiar, acc,prec,rec))
  print(confusion_matrix(y,pred))
```

```python
# Realização da classificação com LR, LDA, QDA. Impressão das métricas de desempenho.
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.model_selection import train_test_split

X_train, X_vali, y_train, y_vali = train_test_split(df,y, test_size=0.20, random_state=40)

clf_lr  = LogisticRegression(solver='newton-cg').fit(X_train,y_train)
clf_lda = LinearDiscriminantAnalysis().fit(X_train,y_train) 
clf_qda = QuadraticDiscriminantAnalysis().fit(X_train,y_train) 

limiar = 0.25
print("LR")
evaluate_model(clf_lr,  X_vali, y_vali, limiar)
print("LDA")
evaluate_model(clf_lda, X_vali, y_vali, limiar)
print("QDA")
evaluate_model(clf_qda, X_vali, y_vali, limiar)

print("-----------------------------")
limiar = 0.50
print("LR")
evaluate_model(clf_lr,  X_vali, y_vali, limiar)
print("LDA")
evaluate_model(clf_lda, X_vali, y_vali, limiar)
print("QDA")
evaluate_model(clf_qda, X_vali, y_vali, limiar)

print("-----------------------------")
limiar = 0.75
print("LR")
evaluate_model(clf_lr,  X_vali, y_vali, limiar)
print("LDA")
evaluate_model(clf_lda, X_vali, y_vali, limiar)
print("QDA")
evaluate_model(clf_qda, X_vali, y_vali, limiar)

```

**Discussão:** A mudança dos limiares não trouxe diferenças tão expressivas. Ao aumentarmos o limiar, vemos um pequeno aumento na precisão para o LDA e um pequeno decremento na cobertura para o LR e QDA. Este comportamento segue o esperado, pois com um limiar maior, o modelo fica mais seletivo ao determinar um resultado como positivo.

Na matriz de confusão estes resultados ficam mais evidentes, com mais Falsos Negativos à medida que aumentamos o limiar. Já os Falsos Positivos praticamente não aumentam quando diminuímos o limiar.

*Comentário extra:* Este é um problema onde o classificador possui um excelente desempenho em todos os casos. A causa provável para a pouca mudança é que a probabilidade dada pelo classificador à classe positiva, quando a resposta correta é positiva, é alta. Já no caso contrário, ela é muito baixa. Assim, com limiares 0.25 e 0.75 os resultados não mudaram muito. 

Quem quiser, pode fazer o teste com limiares 0.05 e 0.95. Vocês vrão que mesmo nesse caso, a precisão e cobertura continuam com valores altos.

```python
# Criação da curva ROC

X_train, X_vali, y_train, y_vali = train_test_split(df,y, test_size=0.20, random_state=40)

clf_lr  = LogisticRegression(solver='newton-cg').fit(X_train,y_train)
clf_lda = LinearDiscriminantAnalysis().fit(X_train,y_train) 
clf_qda = QuadraticDiscriminantAnalysis().fit(X_train,y_train) 

for model in [clf_lr, clf_lda, clf_qda]:
  tpr_list = [] 
  fpr_list = [] 
  pred_proba = model.predict_proba(X_vali)

  for i in range(0,101,1):
      limiar = i/100    
      pred       = (pred_proba[:,1]>limiar)*1
      conf = confusion_matrix(y_vali, pred)
      # TN FP
      # FN TP
      tpr_list.append( conf[1][1] / (conf[1][0]+conf[1][1]) )
      fpr_list.append( conf[0][1] / (conf[0][0]+conf[0][1]) )

  plt.plot(fpr_list, tpr_list,".-")
  plt.show()
  plt.close()

  fpr, tpr, limiar = roc_curve(y_vali, pred_proba[:,1])
  metrica_auc = auc(fpr, tpr)
  print(f'AUC={metrica_auc}')
```
