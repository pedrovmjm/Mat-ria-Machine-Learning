### Exercício 1.3 - [1 ponto]

Veremos agora sobre modos de avaliar o desempenho de um modelo. Suponha que você treinou um modelo de regressão e um de classificação sobre um conjunto de dados de treino. Ao realizar os testes, obteve os seguintes resultados, onde Y é o valor alvo, e Modelo é a saída do seu modelo:

Regressão:

Observação |  Y (alvo)  | Saída do Modelo
-----------|------------|-------------------
1          |   3.0     | 4.0
2          |   5.0     | 7.0
3          |   1.0     | 4.0
4          |   -2.0    | -5.0
5          |   10.0    | 5.0

Classificação:

Observação |  Y (alvo)  | Saída do Modelo
-----------| -------|-------------------
1          |   azul     | vermelho
2          |   azul     | azul
3          |   vermelho | azul
4          |   azul     | azul
5          |   vermelho | vermelho







a) Qual o erro (MSE) do modelo no caso da regressão para cada entrada?

Indique quais as contas realizadas. Exemplo: $\sqrt{2^2 + 3^2} = 4$

**Resposta:**

$MSE = \frac{1}{N} \sum\limits_{i=1}^{N}(Y_{true} - Y_{pred})^2$

```python
import numpy as np

MSE = 0

Y_true = [3, 5, 1, -2, 10]
Y_pred = [4, 7, 4, -5, 5]

N = len(Y_true)

for i in range(N):
  MSE = MSE + np.square(np.subtract(Y_true[i],Y_pred[i]))

MSE = MSE/N

print(f'MSE: {MSE}')
```

***Exato. Se o aluno quiser fazer a conta ao invés de rodar um algoritmo, está ok também. ***

b) Qual o erro (Taxa de acerto) do modelo no caso da classificação para cada entrada?

Indique quais as contas realizadas.

**Resposta:**

**Taxa de acertos:**

$TA = \frac{1}{N} \sum\limits_{i=1}^{N}I(Y_{true} = Y_{pred})$

```python
Y_true = ['azul', 'azul', 'vermelho', 'azul', 'vermelho']
Y_pred = ['vermelho', 'azul', 'azul', 'azul', 'vermelho']

N = len(Y_true)

erros = 0

for i in range(N):
  if Y_true[i] == Y_pred[i]:
    erros = erros + 1
    
taxa = erros/N
print(f'Taxa de acertos: {taxa}')
```

**Obs:** Para os estudantes que calcularam a taxa de erros e não a taxa de acertos por conta do enunciado. Considerem a seguinte resolução:

**Taxa de erros:**

$TE = \frac{1}{N} \sum\limits_{i=1}^{N}I(Y_{true} \neq Y_{pred})$

```python
Y_true = ['azul', 'azul', 'vermelho', 'azul', 'vermelho']
Y_pred = ['vermelho', 'azul', 'azul', 'azul', 'vermelho']

N = len(Y_true)

erros = 0

for i in range(N):
  if Y_true[i] != Y_pred[i]:
    erros = erros + 1
    
taxa = erros/N
print(f'Taxa de erros: {taxa}')
```
